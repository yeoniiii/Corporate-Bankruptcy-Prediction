# 0. 라이브러리

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

# 1. 정상기업

```python
list_corp = pd.read_csv('./data/정상기업.csv', dtype={'종목코드':str})
print(list_corp.shape)
list_corp.head(1)
```

```python
list_fin1 = pd.read_csv('./data/정상기업재무.csv')
print(list_fin1.shape)
list_fin1.head(1)
```

```python
list_fin2 = pd.read_csv('./data/정상기업재무2.csv', dtype={'종목코드':str})
print(list_fin2.shape)
list_fin2.head(1)
```

```python
list_fin1 = pd.merge(list_corp, list_fin1, on='기업명', how='right')
print(list_fin1.shape)
list_fin1.head(3)
```

```python
list_fin2 = pd.merge(list_corp, list_fin2, on='종목코드', how='right')
print(list_fin2.shape)
list_fin2.head(3)
```

```python
list_fin = pd.concat([list_fin1, list_fin2])
print(list_fin.shape)
```

# 2. 부도기업

```python
delist_corp = pd.read_csv('./data/부도기업.csv', dtype={'종목코드':str})
print(delist_corp.shape)
delist_corp.head(1)
```

```python
delist_fin1 = pd.read_csv('./data/부도기업재무.csv', dtype={'종목코드':str})
print(delist_fin1.shape)
delist_fin1.head(1)
```

```python
delist_fin2 = pd.read_csv('./data/부도기업재무2.csv', dtype={'종목코드':str, '연도':int})
delist_fin2['연도'] = delist_fin2['연도'] - 1
print(delist_fin2.shape)
delist_fin2.head(1)
```

```python
delist_fin1 = pd.merge(delist_corp, delist_fin1, on='종목코드', how='right')
print(delist_fin1.shape)
delist_fin1.head(3)
```

```python
delist_fin2 = pd.merge(delist_corp, delist_fin2, on='종목코드', how='right')
print(delist_fin2.shape)
delist_fin2.head(3)
```

```python
delist_fin = pd.concat([delist_fin1, delist_fin2])
print(delist_fin.shape)
```

# 3. 데이터

```python
print(list_fin.shape)
list_fin.head(2)
```

```python
print(delist_fin.shape)
delist_fin.head(2)
```

# 4. 데이터 타입 정리

```python
list_fin.상장일 = pd.to_datetime(list_fin.상장일)
list_fin.지속기간 = pd.to_timedelta(list_fin.지속기간)
list_fin.연도 = list_fin.연도.astype(int)
```

```python
list_fin.dtypes.head()
```

```python
delist_fin.상장일 = pd.to_datetime(delist_fin.상장일)
delist_fin.폐지일 = pd.to_datetime(delist_fin.폐지일)
delist_fin.지속기간 = pd.to_timedelta(delist_fin.지속기간)
delist_fin.연도 = delist_fin.연도.astype(int)
```

```python
delist_fin.dtypes.head()
```

```python
list_fin.rename(columns={'연도':'기준연도'}, inplace=True)
delist_fin.rename(columns={'연도':'기준연도'}, inplace=True)
```

# 5. inf, -inf 값 처리

```python
list_fin[list_fin.isin([np.inf, -np.inf]).any(axis=1)].shape,\
delist_fin[delist_fin.isin([np.inf, -np.inf]).any(axis=1)].shape
```

```python
list_fin.replace([np.inf, -np.inf], np.nan, inplace=True)
delist_fin.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
list_fin[list_fin.isin([np.inf, -np.inf]).any(axis=1)].shape,\
delist_fin[delist_fin.isin([np.inf, -np.inf]).any(axis=1)].shape
```

# 6. 결측치 처리 KNN Imputer

## 1) 정상기업

```python
list_fin.isnull().sum()[list_fin.isnull().sum()!=0]
```

```python
from sklearn.impute import KNNImputer

null_cols = list_fin.iloc[:, 5:].columns

imputer=KNNImputer(n_neighbors=5)
filled_list_fin = imputer.fit_transform(list_fin.iloc[:, 5:])
filled_list_fin = pd.DataFrame(filled_list_fin, columns=list_fin.iloc[:, 5:].columns)
```

```python
filled_list_fin.shape, list_fin.shape
```

```python
list_fin.iloc[:, 5:] = filled_list_fin.copy()
list_fin.head(2)
```

```python
list_fin.isnull().sum()[list_fin.isnull().sum()!=0]
```

## 2) 부도기업

```python
delist_fin.isnull().sum()[delist_fin.isnull().sum()!=0]
```

```python
from sklearn.impute import KNNImputer

null_cols = delist_fin.iloc[:, 6:].columns

imputer=KNNImputer(n_neighbors=5)
filled_delist_fin = imputer.fit_transform(delist_fin.iloc[:, 6:])
filled_delist_fin = pd.DataFrame(filled_delist_fin, columns=delist_fin.iloc[:, 6:].columns)
```

```python
delist_fin.iloc[:, 6:] = filled_delist_fin.copy()
delist_fin.head(2)
```

```python
delist_fin.isnull().sum()[delist_fin.isnull().sum()!=0]
```
